```python
def remove_non_numeric_symbols(df):
  df['Price per Unit'] = df['Price per Unit'].astype(str).str.replace(',', '')
  df['Price per Unit'] = df['Price per Unit'].replace('$', '').astype(float)

  df['Units Sold'] = df['Units Sold'].astype(str).str.replace(',', '')
  df['Units Sold'] = df['Units Sold'].astype(str).str.replace('$', '').astype(float)

  df['Total Sales'] = df['Total Sales'].astype(str).str.replace(',', '').str.strip()
  df['Total Sales'] = df['Total Sales'].astype(str).str.replace('$', '').astype(float)

  df['Operating Profit'] = df['Operating Profit'].astype(str).str.replace(',', '')
  df['Operating Profit'] = df['Operating Profit'].astype(str).str.replace('$', '').astype(float)

  df['Operating Margin'] = df['Operating Margin'].astype(str).str.replace(',', '')
  df['Operating Margin'] = df['Operating Margin'].astype(str).str.replace('%', '').astype(float)


remove_non_numeric_symbols(df)
```

```python
df.head()
```

extrcting year & month

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

df['Year'] = df['Invoice Date'].dt.year
df['Month'] = df['Invoice Date'].dt.month
```

```python
df.head()
```

**FINANCIAL PERFORMANCE ANALYSIS**  
* total revenue
* revenue for each year  
* revenue for each month
* revenue same period last year

***TOTAL REVENUE***

```python
total_revenue = sum(df['Total Sales'])
total_revenue
```

***REVENUE FOR EACH YEAR***

```python
revenue_for_each_year = df.groupby(by = df['Year'])['Total Sales'].sum()
print(revenue_for_each_year)
```

```python
revenue_for_each_year.plot(kind = 'bar')
plt.legend()
```

***REVENUE FOR EACH MONTH***

```python
revenue_for_each_month = df.groupby(by = df['Month'])['Total Sales'].sum()
print(revenue_for_each_month)
```

```python
revenue_for_each_month.plot()
plt.legend()
```

**REVENUE SAME PERIOD LAST YEAR**

```python
#Calculate revenue for same period last year
revenue_last_year = []
for index, row in df.iterrows():
    same_period_last_year = df[(df['Year'] == row['Year'] - 1) & (df['Month'] == row['Month'])]
    revenue_last_year.append(same_period_last_year['Total Sales'].sum())

df['Revenue Last Year'] = revenue_last_year
```
